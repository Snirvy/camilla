```python
ulist = []
glist = []
dlist = []
flist = []
genlist = []
devlist = []
transcriptlength = []

for file in os.listdir():
    
    

    with open(file) as current:
        f = str(current.readlines())
        
        textlen = len(f)
        
        if 'adhd' in file:
            development = 'ADHD'   
        else:
            development = 'TD'

        if 'm' in file:
            gender = 'M'  
        else:
            gender = 'F'

        u = "Unfinished"
        ucount = 0

        g = "Grammatical"
        gcount = 0

        d = "Dialect"
        dcount = 0

        fi = "Fillers"
        fcount = 0

        data = f.strip().split()

        for i in range(len(data)):


            if u in data[i]:
                ucount += 1
            if g in data[i]:
                gcount += 1
            if d in data[i]:
                dcount += 1
            if fi in data[i]:
                fcount += 1

        ulist.append(ucount)
        glist.append(gcount)
        dlist.append(dcount)
        flist.append(fcount)
        genlist.append(gender)
        devlist.append(development)
        transcriptlength.append(textlen)
 
#     print("Unfinished Utterances: {}".format(ucount))

#     print("Grammatical Errors: {}".format(gcount))

#     print("Dialect Uses: {}".format(dcount))

#     print("Filler Uses: {}".format(fcount))



    
```

```python
def cleanstring( s ):
    news = ""
    s = s.lower()
    for c in s:
        if c >= "a" and c <= "z":
            news += c
        else:
            news += " "
    return news
```
